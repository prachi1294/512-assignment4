```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt

metrics = pd.read_csv('metrics.csv', index_col=0)
metrics
```

```python
metrics2 = pd.read_csv('metrics.ppn2.csv', index_col=0)
plt.figure()
plt.plot(metrics.elapsed_time, metrics.epoch_losses, label='PPN 10')
plt.plot(metrics2.elapsed_time, metrics2.epoch_losses, label='PPN 2')
plt.xlabel('Elapsed Time')
plt.ylabel('Training Losss')
plt.legend()
```

```python
plt.figure()
plt.plot(metrics.epoch_losses, label='PPN 10')
plt.plot(metrics2.epoch_losses, label='PPN 2')
plt.xlabel('Epoch')
plt.ylabel('Training Loss')
plt.legend()
```
