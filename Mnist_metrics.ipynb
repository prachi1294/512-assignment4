```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt

metrics_1 = pd.read_csv('metrics_1.csv', index_col=0)
metrics_2 = pd.read_csv('metrics_2.csv', index_col=0)
metrics_4 = pd.read_csv('metrics_4.csv', index_col=0)
metrics_8 = pd.read_csv('metrics_8.csv', index_col=0)
metrics_2
```

```python
time = [metrics_1['elapsed_time'], metrics_2['elapsed_time'], metrics_4['elapsed_time'], metrics_8['elapsed_time']]
accuracy = [metrics_1['accuracies'], metrics_2['accuracies'], metrics_4['accuracies'], metrics_8['accuracies']]
labels = ['process 1', 'process 2', 'process 4', 'process 8']
```

```python
plt.figure()
for i in range(len(accuracy)):
    
    plt.plot(np.arange(30), accuracy[i], label = labels[i])

    
plt.title('Accuracy vs. Epoch with four processes')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
plt.figure()
for i in range(len(time)):
    
    plt.plot(np.arange(30), time[i], label = labels[i])

    
plt.title('Elapsed Time vs. Epoch with four processes')
plt.xlabel('Epoch')
plt.ylabel('Time(s)')
plt.legend()
plt.show()
```

 ## Effect of Training time
 It is noticed that training time is decreases as we increase the number of process. However, I have reduced the batch size to 32 as it wasn't working for 8 processes but still it was fatster. Althought, we can see a speedup where 30 epochs takes almost above 600 with one process, 350 sec with 2 processes, 150 sec with 4 processes and 100 sec with 8 processes. so it's seems like Gustafson-Barsis law could be a best.
 
 ## Effect on Accuracy
 We can see that accuracy is decreasing because of not using the entire dataset? Also, we can note the 'bumps' on line while using 8 processes. We could get good accuracy even with 8 processes but in that case we have to increase no of epochs and dataset size.
